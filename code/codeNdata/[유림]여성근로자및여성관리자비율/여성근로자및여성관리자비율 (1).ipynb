```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
import random
import platform

data = pd.read_csv('여성근로자_및_여성관리자_비율_20220120185821.csv',encoding = 'cp949' )
```

```python
def os_setting():
    if platform.system() == 'Darwin': #맥
            plt.rc('font', family='AppleGothic') 
    elif platform.system() == 'Windows': #윈도우
            plt.rc('font', family='Malgun Gothic') 
    elif platform.system() == 'Linux': #리눅스 (구글 콜랩)
            #!wget "https://www.wfonts.com/download/data/2016/06/13/malgun-gothic/malgun.ttf"
            #!mv malgun.ttf /usr/share/fonts/truetype/
            #import matplotlib.font_manager as fm 
            #fm._rebuild() 
            plt.rc('font', family='Malgun Gothic') 
    plt.rcParams['axes.unicode_minus'] = False #한글 폰트 사용시 마이너스 폰트 깨짐 해결
    
os_setting()
```

```python
df = data[['2020','2020.1','2020.2','2020.3','2020.4','2020.5','2020.6','2020.7']]
df = df.iloc[:3,:]
df.iloc[2,:] = df.iloc[2,:].apply(pd.to_numeric)
```

```python
df=df.T
```

```python
df.columns = ['분류', '기업', '비율']
```

```python
df_pivot = df.pivot(index = '기업', columns='분류', values='비율')
```

```python
df_pivot.plot.bar(figsize=(50,25), color = ['lightskyblue', 'lightcoral'])
plt.rc('font', size=25) 
plt.rc('axes', labelsize=50)   # x,y축 label 폰트 크기
plt.rc('xtick', labelsize=50)  # x축 눈금 폰트 크기 
plt.rc('ytick', labelsize=50)  # y축 눈금 폰트 크기
plt.rc('legend', fontsize=50)  # 범례 폰트 크기
plt.rc('figure', titlesize=50) # figure title 폰트 크기
plt.xticks(rotation=0)
plt.title('여성근로자 및 여성관리자비율', fontsize=50) 

plt.savefig('여성근로자 및 여성관리자비율.png')
```
