```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
import os, glob
import os.path
import matplotlib.gridspec as gridspec
import random
from matplotlib import rc
%matplotlib inline
plt.rcParams['axes.unicode_minus'] = False
import platform
def os_setting():
    if platform.system() == 'Darwin': #맥
            plt.rc('font', family='AppleGothic') 
    elif platform.system() == 'Windows': #윈도우
            plt.rc('font', family='Malgun Gothic') 
    elif platform.system() == 'Linux': #리눅스 (구글 콜랩)
            #!wget "https://www.wfonts.com/download/data/2016/06/13/malgun-gothic/malgun.ttf"
            #!mv malgun.ttf /usr/share/fonts/truetype/
            #import matplotlib.font_manager as fm 
            #fm._rebuild() 
            plt.rc('font', family='Malgun Gothic') 
    plt.rcParams['axes.unicode_minus'] = False #한글 폰트 사용시 마이너스 폰트 깨짐 해결
    
os_setting()

data8_df=pd.read_csv('data8.csv')
df=data8_df[['시도명','시군구명']]
```

시도별

```python
freq=df['시도명'].value_counts()
df1=pd.DataFrame(freq)
df1.reset_index(drop=False, inplace=True)
df1.columns=['시도명','개수']
arrName=df1['시도명'].values
arrNum=df1['개수'].values
explode=[0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.12]
plt.figure(figsize=(7,7))
plt.pie(arrNum,labels=arrName,explode=explode,startangle=90,shadow=False,autopct='%.1f',colors=['orange','gold','cornsilk','goldenrod','wheat','yellow','yellowgreen','honeydew','lightgreen','olive','forestgreen','limegreen','mediumseagreen','turquoise','paleturquoise','cadetblue','skyblue'])
plt.title("여성새로일하기센터 시도별",fontsize=30)
plt.show()
```
