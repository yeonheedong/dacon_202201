os_setting()

data3_df=pd.read_csv('data3.csv')
df=data3_df[['성별','산업분류','2020','2020.4']]
df=df.drop(index=[0,1,20,39],axis=0)
df = df.reset_index()
del(df['index'])
df.columns=['성별','산업분류','초임평균','전체평균']

**초임평균**

```python
df_pivot1 = df.pivot(index = '산업분류', columns='성별', values='초임평균')
df_pivot1['남자']=df_pivot1['남자'].astype(float)
df_pivot1['여자']=df_pivot1['여자'].astype(float)
df_pivot1['전체']=df_pivot1['전체'].astype(float)

df_pivot1.plot.bar(figsize=(50,25), color = ['cornflowerblue', 'lightcoral', 'yellowgreen'])
plt.rc('font', size=25) 
plt.rc('axes', labelsize=50)   # x,y축 label 폰트 크기
plt.rc('xtick', labelsize=50)  # x축 눈금 폰트 크기 
plt.rc('ytick', labelsize=50)  # y축 눈금 폰트 크기
plt.rc('legend', fontsize=50)  # 범례 폰트 크기
plt.rc('figure', titlesize=50) # figure title 폰트 크기
#path='/content/drive/MyDrive/'
#plt.savefig(path + '초임평균.jpg')
```

전체평균

```python
df_pivot2 = df.pivot(index = '산업분류', columns='성별', values='전체평균')
df_pivot2['남자']=df_pivot2['남자'].astype(float)
df_pivot2['여자']=df_pivot2['여자'].astype(float)
df_pivot2['전체']=df_pivot2['전체'].astype(float)

df_pivot2.plot.bar(figsize=(50,25), color = ['cornflowerblue', 'lightcoral', 'yellowgreen'])
plt.rc('font', size=25) 
plt.rc('axes', labelsize=50)   # x,y축 label 폰트 크기
plt.rc('xtick', labelsize=50)  # x축 눈금 폰트 크기 
plt.rc('ytick', labelsize=50)  # y축 눈금 폰트 크기
plt.rc('legend', fontsize=50)  # 범례 폰트 크기
plt.rc('figure', titlesize=50) # figure title 폰트 크기
#path='/content/drive/MyDrive/'
#plt.savefig(path + '전체평균.jpg')
```
