```python
# 선그래프
fig = plt.figure(figsize=(10,10)) ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 색상 설정
ax = fig.add_subplot() ## 그림 뼈대(프레임) 생성


plt.plot(df.index, df['가사부담'],marker='s',color='r', label='가사부담')
plt.plot(df.index, df['불평등한 근로여건'],marker='o',color='g',label='불평등한 근로여건')
plt.plot(df.index, df['여성에 대한 사회적 편견'],marker='*',color='b',label='여성에 대한 사회적 편견')
plt.plot(df.index, df['육아부담'],marker='+',color='y',label='육아부담')

plt.title('여성의 취업 장애 요인(2009-2021)', fontsize=20) 
plt.ylabel('비율', fontsize=14)
plt.xlabel('연도', fontsize=14)
plt.legend(fontsize=12, loc='best')
plt.show()
```
