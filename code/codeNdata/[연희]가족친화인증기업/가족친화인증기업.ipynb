```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
import os, glob
import os.path
import matplotlib.gridspec as gridspec
import random
from matplotlib import rc
%matplotlib inline
plt.rcParams['axes.unicode_minus'] = False
import platform
def os_setting():
    if platform.system() == 'Darwin': #맥
            plt.rc('font', family='AppleGothic') 
    elif platform.system() == 'Windows': #윈도우
            plt.rc('font', family='Malgun Gothic') 
    elif platform.system() == 'Linux': #리눅스 (구글 콜랩)
            #!wget "https://www.wfonts.com/download/data/2016/06/13/malgun-gothic/malgun.ttf"
            #!mv malgun.ttf /usr/share/fonts/truetype/
            #import matplotlib.font_manager as fm 
            #fm._rebuild() 
            plt.rc('font', family='Malgun Gothic') 
    plt.rcParams['axes.unicode_minus'] = False #한글 폰트 사용시 마이너스 폰트 깨짐 해결
    
os_setting()

data9_df=pd.read_csv('data9.csv',encoding='cp949')
df1=data9_df[['구분','분류']]
```

분류별(2018)

```python
freq1=df1['분류'].value_counts()
df1=pd.DataFrame(freq1)
df1.reset_index(drop=False, inplace=True)
df1.columns=['분류','개수']
arrName1=df1['분류'].values
arrNum1=df1['개수'].values
explode=[0.05,0.05,0.05]
plt.figure(figsize=(7,7))
plt.pie(arrNum1,labels=arrName1,explode=explode,startangle=90,shadow=False,autopct='%.1f',colors=['gold','wheat','olive'])
plt.title("가족친화인증기업 기업분류별",fontsize=30)
plt.show()
```

시도별(2018)

```python
df2=data9_df[['구분','시도']]
freq2=df2['시도'].value_counts()
df2=pd.DataFrame(freq2)
df2.reset_index(drop=False, inplace=True)
df2.columns=['시도','개수']
arrName2=df2['시도'].values
arrNum2=df2['개수'].values
explode=explode=[0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05]
plt.figure(figsize=(7,7))
plt.pie(arrNum2,labels=arrName2,explode=explode,startangle=90,shadow=False,autopct='%.1f',colors=['orange','gold','cornsilk','goldenrod','wheat','yellow','yellowgreen','honeydew','lightgreen','olive','forestgreen','limegreen','mediumseagreen','turquoise','paleturquoise','cadetblue','skyblue'])
plt.title("가족친화인증기업 시도별",fontsize=30)
plt.show()
```
